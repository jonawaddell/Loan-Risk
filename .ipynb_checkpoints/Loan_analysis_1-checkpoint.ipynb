```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
loan_data_file = "loans_data6.csv"
```

```python
loan_data_pd = pd.read_csv(loan_data_file)
loan_data_pd.head()
```

```python
organized_loanpurpose_pd = loan_data_pd[['loan_amnt', 'purpose']]
organized_loanpurpose_pd.head()
```

```python
organized_loanpurpose_pd.columns
```

```python
debt_consolidation_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='debt_consolidation']
credit_card_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='credit_card'] 
medical_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='medical'] 
other_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='other'] 
vacation_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='vacation'] 
majorpurchase_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='major_purchase'] 
house_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='house'] 
smallBusiness_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='small_business'] 
homeImprovement_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='home_improvemnt'] 
car_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='car'] 
moving_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='moving'] 
renew_energy_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='renewable_energy'] 
educational_pd = organized_loanpurpose_pd.loc[organized_loanpurpose_pd['purpose']=='educational']
```

```python
plt.figure(figsize = (15,15))
plt.plot(np.arange(len(debt_consolidation_pd['loan_amnt'])), debt_consolidation_pd['loan_amnt'], marker = 'o',color = "r", linewidth=0.5, alpha=0.8, label = "debt_consolidation")
plt.plot(np.arange(len(medical_pd['loan_amnt'])), medical_pd['loan_amnt'], marker = 'o',color = "m", linewidth=0.5, alpha=0.8, label = "medical")
plt.plot(np.arange(len(credit_card_pd['loan_amnt'])), credit_card_pd['loan_amnt'], marker = 'o',color = "b", linewidth=0.5, alpha=0.8, label = "credit_card")
plt.plot(np.arange(len(other_pd['loan_amnt'])), other_pd['loan_amnt'], marker = 'o',color = "y", linewidth=0.5, alpha=0.8, label = "other")
plt.plot(np.arange(len(vacation_pd['loan_amnt'])), vacation_pd['loan_amnt'], marker = 'o',color = "k", linewidth=0.5, alpha=0.8, label = "vacation")
plt.plot(np.arange(len(majorpurchase_pd['loan_amnt'])), majorpurchase_pd['loan_amnt'], marker = 'o',color = "g", linewidth=0.5, alpha=0.8, label = "majorpurchase_pd")
plt.plot(np.arange(len(house_pd['loan_amnt'])), house_pd['loan_amnt'], marker = 'o',color = "r", linewidth=0.5, alpha=0.8, label = "house")
plt.plot(np.arange(len(smallBusiness_pd['loan_amnt'])), smallBusiness_pd['loan_amnt'], marker = 'o',color = "m", linewidth=0.5, alpha=0.8, label = "smallBusiness")
plt.plot(np.arange(len(homeImprovement_pd['loan_amnt'])), homeImprovement_pd['loan_amnt'], marker = 'o',color = "b", linewidth=0.5, alpha=0.8, label = "homeImprovement")
plt.plot(np.arange(len(car_pd['loan_amnt'])), car_pd['loan_amnt'], marker = 'o',color = "k", linewidth=0.5, alpha=0.8, label = "car")
plt.plot(np.arange(len(moving_pd['loan_amnt'])), moving_pd['loan_amnt'], marker = 'o',color = "m", linewidth=0.5, alpha=0.8, label = "moving")
plt.plot(np.arange(len(renew_energy_pd['loan_amnt'])), renew_energy_pd['loan_amnt'], marker = 'o',color = "g", linewidth=0.5, alpha=0.8, label = "renew_energy")
plt.plot(np.arange(len(educational_pd['loan_amnt'])), educational_pd['loan_amnt'], marker = 'o',color = "g", linewidth=0.5, alpha=0.8, label = "education")

plt.xlim([0, 65000])
plt.grid()
plt.show()
```
